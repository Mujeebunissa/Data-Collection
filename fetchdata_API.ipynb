```python
import requests

API_KEY = "1602de9771a7b7289abdfaa7e1243acc"
BASE_URL = "https://api.themoviedb.org/3"
TOP_RATED_MOVIES_ENDPOINT = "/movie/top_rated"

# Construct the URL
url = f"{BASE_URL}{TOP_RATED_MOVIES_ENDPOINT}?api_key={API_KEY}"
print(url)

# Send GET request to the API
response = requests.get(url)
print(response)                       

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract URLs of top-rated movies
    top_rated_movies_urls = []
    for movie in data['results']:
        movie_id = movie['id']
        movie_url = f"https://www.themoviedb.org/movie/{movie_id}"
        top_rated_movies_urls.append(movie_url)
    
    # Print or use the URLs as needed
    for url in top_rated_movies_urls:
        print(url)
else:
    print("Error:", response.status_code)
```

```python
import requests
import pandas as pd

# Create an empty list to store the results
results_list = []

# Iterate over the range of pages
for i in range(1, 461):  # Range should start from 1, not 0
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=1602de9771a7b7289abdfaa7e1243acc&language=en-US&page={i}')
    
    # Check if the request was successful
    if response.status_code == 200:
        # Extract data from the response and append it to the list
        results_list.extend(response.json().get('results', []))
    else:
        print(f"Failed to fetch page {i}. Status code: {response.status_code}")

# Create a DataFrame from the list of results
df = pd.DataFrame(results_list)

# Filter the DataFrame to include only desired columns
df = df[['id', 'title', 'overview', 'popularity', 'vote_average', 'vote_count']]

# Display the DataFrame
display(df)
```
